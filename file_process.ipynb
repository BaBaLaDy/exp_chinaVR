```python
# merge_fold

import os
import csv

def merge_csv_files(parent_directory, prefix, output_file):
    # 获取所有以prefix开头的文件夹
    folders = [f for f in os.listdir(parent_directory) if os.path.isdir(os.path.join(parent_directory, f)) and f.startswith(prefix)]
    print(folders)
    
    # 打开输出文件准备写入
    with open(output_file, 'w', newline='', encoding='utf-8') as outfile:
        writer = csv.writer(outfile)
        
        # 遍历每个文件夹
        for folder in folders:
            folder_path = os.path.join(parent_directory, folder)
            
            # 遍历每个文件夹中的csv文件
            for file in os.listdir(folder_path):
                if file.endswith('.csv'):
                    file_path = os.path.join(folder_path, file)
                    # 读取csv文件，并只保留前5列
                    with open(file_path, 'r', encoding='utf-8') as infile:
                        reader = csv.reader(infile)
                        for row in reader:
                            writer.writerow(row)

# 使用示例
parent_directory = 'D:\save_data\SMPL_unity_v.1.0.0\smpl_mecanim\\assets\\0713Generated_ChinaVR\WALK\\'

motion = 'WALK/'
prefix = 'WALK3_'  # 例如 'prefix1', 'prefix2' 等
save_data_file = 'datasets/Merge/' + motion + prefix + '.csv'
if not os.path.exists('datasets/Merge/' + motion):
    os.makedirs('datasets/Merge/' + motion)

merge_csv_files(parent_directory, prefix, save_data_file)
# combined_df.to_csv('combined_output.csv', index=False)
```
